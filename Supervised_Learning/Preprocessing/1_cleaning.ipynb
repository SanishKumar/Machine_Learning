```python
#import csv

df = pd.read_csv('sanish.csv')
```

```python
#rename multiple columns in a list

df = df.rename(columns = {'Mat':'Matches','NO':'Not_Outs', 'HS':'Highest_Inns_Score', 'BF':'Balls_Faced','SR':'Strike_Rate'})
```

```python
df = df.drop_duplicates()
```

```python
#Split up Span into Start and End Date

df['Span'].str.split(pat = '-')
```

```python
df['Rookie_Year'] = df['Span'].str.split(pat = '-').str[0]
```

```python
df['Final_Year'] = df['Span'].str.split(pat = '-').str[1]
```

```python
df
```

```python
df = df.drop(['Span'], axis = 1)
```

```python
df.head()
```

```python
#Split the Country with Player

# df['Player'].str.split(pat = '(')

df['Player_Name'] = df['Player'].str.split(pat="(").str[0]
```

```python
df
```

```python
df['Country'] = df['Player'].str.split(pat="(").str[1]
df['Country'] = df['Country'].str.replace(")","")
```

```python
df['Player'] = df['Player_Name']
df = df.drop(['Player_Name'],axis=1)
df
```

```python
df.dtypes
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.split(pat='*').str[0]
df['Matches'] = df['Matches'].str.split(pat='*').str[0]
```

```python
df
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype('int64')
df['Matches '] = df['Matches '].astype('int64')
```

```python
df.dtypes
```

```python
df['4s'] = df['4s'].str.split(pat='+').str[0]
df['4s'] = df['4s'].astype('int64')
df['6s'] = df['6s'].str.split(pat='+').str[0]
df['6s'] = df['6s'].astype('int64')
df['Balls_Faced'] = df['Balls_Faced'].str.split(pat='+').str[0]

# Fill NaN values with 0 or any other appropriate value
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)

# Now convert to integer
df['Balls_Faced'] = df['Balls_Faced'].astype('int64')

df = df.astype({'Rookie_Year':'int64','Final_Year':'int64'})
```

```python
df.dtypes
```

```python
df['Career_length'] = df['Final_Year'] - df['Rookie_Year']
```
